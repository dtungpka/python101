````python
import pickle
# B·∫ÆT ƒê·∫¶U CODE
MSSV = "2xxxxxxx"
# K·∫æT TH√öC CODE
get_list = ['Sin','Cos','Data','Thoi_gian','Nam','Nam_CC','nam_2018','nu_2018','dientich','danso']
glb_vars = vars()
with open(str(MSSV)+'.kq5', 'wb') as handle:
    g = dict()
    for i in glb_vars:
      if i in get_list:
        g[i] = glb_vars[i]
    pickle.dump(g, handle, protocol=pickle.HIGHEST_PROTOCOL)
    handle.close()
````
