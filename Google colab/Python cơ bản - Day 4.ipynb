####Bài 1

Cho sẵn mảng arr, thực hiện:

*   Chuyển arr thành ma trận numpy A,
*   Tính A nhân 2
*   Chuyển ma trận A thành ma trận 3x3


 
 ma trận mong muốn:
 ```python
 A = [[2,4,6],
      [8,10,12]
      [14,16,18]]
 ```
 


````python
import numpy as np
arr = [1,2,3,4,5,6,7,8,9]


#(≈ 3 dòng code)
# BẮT ĐẦU CODE
A = np.array(arr)
A = A * 2
A = A.reshape(3,3)
# KẾT THÚC CODE
print(A)
````

####Bài 2

Cho bảng sau, tính các phép tính và lưu thành 1 ma trận với tên đặt lần lượt là  ans1,ans2,ans3,ans4, với ans1 là ma trận 1x4 chứa kết quả của (X+Y)^2

| **x** | **y** | **z** | **(x + y)^2** | **x^2 + y^2 + 2xy** | **(x + y + z)^3** | **(xy)^2 + (yz)^2 + (xz)^2** |
|-------|-------|-------|---------------|---------------------|-------------------|------------------------------|
| 9     | 5     | 11    |               |                     |                   |                              |
| 125   | 48    | 24    |               |                     |                   |                              |
| 32    | 18    | 32    |               |                     |                   |                              |
| 29    | 12    | 7     |               |                     |                   |                              |



````python
'''
X = np.array([..]).reshape((..,..))
...

ans1 = (X + ...

'''

#(≈ 8 dòng code)
# BẮT ĐẦU CODE
X = np.array([9,125,32,29]).reshape(4,1)
Y = np.array([5,48,18,12]).reshape(4,1)
Z = np.array([11,24,32,7]).reshape(4,1)

ans1 = (X + Y ) ** 2
ans2 = X**2 + Y**2 + 2 * X * Y
ans3 = (X + Y + Z)**3
ans4 = (X*Y)**2 + (Y*Z)**2 + (X*Z)**2




# KẾT THÚC CODE
print(ans1,ans2,ans3,ans4)
````

Lập ma trận tính mỗi cột còn trống trong bảng sau (mỗi cột là một ma trận 1x7),
biết giá điện là 800đ và giá công tơ là 10000đ
Sau khi tính, lưu tiền phải trả (cột cuối) vào dictionary TienDien với `{'tên chủ hộ' : Tiền phải trả} `

Lưu ý: Cột tên chủ hộ có thể dùng mảng của python hoặc ma trận numpy đều được

| STT | Tên chủ hộ | Chỉ số trước | Chỉ số sau | Điện tiêu thụ | Tiền điện | Tiền công tơ | Tiền phải trả |
|-----|------------|--------------|------------|---------------|-----------|--------------|---------------|
| 1   | Vân        | 45           | 450        |               |           |              |               |
| 2   | Bình       | 5            | 570        |               |           |              |               |
| 3   | Lan        | 4            | 546        |               |           |              |               |
| 4   | Khánh      | 78           | 123        |               |           |              |               |
| 5   | Doanh      | 7            | 46         |               |           |              |               |
| 6   | Thu        | 100          | 212        |               |           |              |               |
| 7   | Hường      | 21           | 67         |               |           |              |               |

````python
'''
chiSotruoc = np.array(..).reshape((..,..))
...
...
#Lặp qua tên chủ hộ và cho vào dict cùng với tiền phải trả
TienPhaiTra = {}
for i in len(TenChuHo):
  ten = str( TenChuHo[i] )
  TienPhaiTra[ ten ] = ...
'''
#(≈ 12 dòng code)
# BẮT ĐẦU CODE

Ten = ["Vân","Bình"]
chiSotruoc = np.array([45,5]).reshape(2,1)
chiSosau = np.array([450,570]).reshape(2,1)

tieuthu = chiSosau - chiSotruoc
tiendien = tieuthu * 800
tiencongto = tieuthu * 10000
tienphaitra = tiendien + tiencongto
TienDien = {}
for i in range(len(Ten)):
  ten = Ten[i]
  TienDien[ten] = int(tienphaitra[i])




# KẾT THÚC CODE
for t in TienDien:
  print(f"Hộ {t} hết {TienDien[t]}đ tiền điện") 
````

#Tạo file kết quả
Sau khi bạn làm xong tất cả bài tập bên trên hãy sửa biến MSSV dưới đây từ 2xxxxxxx thành mã số sinh viên của bạn; ví dụ:
`MSSV = "21010294"`
, sau đó chạy đoạn code để tạo file kết quả.

Kết quả của bạn sẽ được lưu thành file có tên là mã sinh viên của bạn, sau đó bạn bấm vào Files > chọn file của bạn > Download để tải file kết quả bài làm về máy.

image.png

````python
import pickle
# BẮT ĐẦU CODE
MSSV = "2xxxxxxx"
# KẾT THÚC CODE
get_list = ['A','ans1','ans2','ans3','ans4','TienDien']
glb_vars = vars()
with open(str(MSSV)+'.kq4', 'wb') as handle:
    g = dict()
    for i in glb_vars:
      if i in get_list:
        g[i] = glb_vars[i]
    pickle.dump(g, handle, protocol=pickle.HIGHEST_PROTOCOL)
    handle.close()
````
